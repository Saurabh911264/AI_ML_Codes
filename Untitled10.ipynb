```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive')
```

```python
dataset = pd.read_csv(r'/content/drive/My Drive/Machine_Learning/ViratKohli.csv')
print(dataset, "\n")
```

```python
dataset = dataset.drop('Strike Rate', axis=1)
```

```python
print(dataset.columns)
import matplotlib.pyplot as plt
```

```python
# List of batting orders (1 to 6)
batting_orders = [1, 2, 3, 4, 5, 6]

# Create a bar graph for each batting order
for order in batting_orders:
    # Filter data for the specific batting order
    filtered_data = dataset[dataset['Batting Order'] == order]

    # Calculate the total runs for each player in that order
    player_runs = filtered_data.groupby('Man of the Match')['Score'].sum()

    # Plot the bar graph
    plt.figure(figsize=(10, 6))
    plt.bar(player_runs.index, player_runs.values)
    plt.xlabel('Player')
    plt.ylabel('Total Runs')
    plt.title(f'Runs vs. Batting Order {order}')
    plt.xticks(rotation=45)
    plt.show()
```
